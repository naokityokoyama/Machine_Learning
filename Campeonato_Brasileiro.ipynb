```
def pontos(clube):
  time = data_12.loc[(data_12['Clube 1']==clube) | (data_12['Clube 2']==clube)]
  time_A = time.loc[(time['Estado Clube Vencedor']=='Empate') | (time['Vencedor']==clube)]
  time_A = list(time_A['Estado Clube Vencedor'].values)

  lista = []
  for i in time_A:
    if i=='Empate':
      lista.append(1)
    else:
      lista.append(3)
  return (sum(lista))
```

```
clubes = list(data_12['Clube 1'].unique())
pontos('Sport')

lista_clubes = []
for i in clubes:
  dic = {}
  dic['Clubes'] = i
  dic['Pontos'] = pontos(i)
  lista_clubes.append(dic)
```
